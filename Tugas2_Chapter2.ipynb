# Chapter 2: Loading Data

##Setup and Library Imports

```python
import pandas as pd
import numpy as np

print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## Loading a CSV File

CSV (*Comma Separated Values*) adalah format file yang paling umum kita temui dalam analisis data. Pandas mempermudah kita untuk memuat file ini menjadi sebuah objek yang disebut **DataFrame**. DataFrame ini bisa kita bayangkan seperti tabel di spreadsheet (Excel), namun dengan kekuatan pemrosesan Python yang jauh lebih cepat.

```python
url_ev = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
df_ev = pd.read_csv(url_ev)

print("Data Kendaraan Listrik (5 baris teratas):")
display(df_ev.head())

print(f"\nTotal data yang dimuat: {len(df_ev)} baris.")
```

## Loading a JSON File

JSON adalah format yang sangat ringan dan sering digunakan untuk pengiriman data antar aplikasi web. Struktur JSON biasanya berbentuk seperti kamus (*dictionary*) di Python dengan pasangan kunci dan nilai (key-value pairs). Pandas memudahkan kita untuk mengubah struktur yang bersifat hierarkis ini menjadi tabel DataFrame yang rapi.

```python
url_json = "https://raw.githubusercontent.com/domoritz/maps/master/data/iris.json"

df_json = pd.read_json(url_json)

print("Data Iris dari format JSON:")
display(df_json.head())
```

## Querying a SQL Database

Di dunia industri, data sering kali tidak disimpan dalam file satuan, melainkan di dalam database. Untuk mengambilnya, kita bisa menggunakan bahasa SQL. Pandas memiliki fungsi `read_sql_query` yang sangat kuat untuk menarik data langsung dari database ke dalam DataFrame kita.

Sebagai contoh, kita akan membuat database sementara di memori menggunakan `sqlite3` untuk mensimulasikan proses ini.

```python
import sqlite3

conn = sqlite3.connect(":memory:")
df_ev.head(100).to_sql("ev_table", conn)
query = "SELECT City, Make, Model FROM ev_table WHERE Make = 'TESLA'"
df_tesla = pd.read_sql_query(query, conn)

print("Hasil Filter SQL (Hanya mobil Tesla):")
display(df_tesla.head())
```

## Describing the Data

Setelah kita berhasil memuat data, langkah selanjutnya adalah melakukan eksplorasi dasar. Kita perlu memahami struktur data, tipe data di setiap kolom, serta statistik deskriptifnya. Ini akan membantu kita mengidentifikasi apakah ada keanehan pada data (seperti nilai yang hilang atau *outlier*) sebelum kita melangkah ke tahap analisis yang lebih dalam.

Kita akan menggunakan tiga fungsi utama dari Pandas untuk "mengintip" data kita:
1. `shape`: Untuk melihat dimensi (jumlah baris dan kolom).
2. `info()`: Untuk melihat tipe data dan memori yang digunakan.
3. `describe()`: Untuk melihat statistik ringkasan seperti rata-rata, nilai minimum, dan maksimum.

```python
print("Dimensi Data (Baris, Kolom):", df_ev.shape)

print("\nInformasi Struktur Data:")
df_ev.info()

print("\nStatistik Deskriptif Data:")
display(df_ev.describe())
```

## Navigating DataFrames

Untuk menavigasi atau mengambil bagian tertentu dari data, kita memiliki dua metode utama di Pandas:

1. **iloc**: Digunakan jika kita ingin mengambil data berdasarkan **posisi indeksnya** (berupa angka). Ini sangat berguna jika kita ingin melakukan *slicing* seperti pada list Python biasa.
2. **loc**: Digunakan jika kita ingin mengambil data berdasarkan **label atau nama kolomnya**.

Dengan menggunakan kedua metode ini, kita bisa memilih baris tunggal, rentang baris, hingga kolom-kolom tertentu dengan sangat fleksibel.

```python
print("Baris pertama (posisi 0):")
display(df_ev.iloc[0])

print("\nBaris ke-2 sampai ke-4:")
display(df_ev.iloc[1:4])

print("\nKolom City dan Make untuk 5 baris pertama:")
display(df_ev.loc[0:4, ['City', 'Make']])
```

## Selecting Rows Based on Conditions

Sering kali kita tidak membutuhkan seluruh isi dataset. Kita mungkin hanya ingin menganalisis data yang memenuhi kriteria tertentu. Di Pandas, kita bisa melakukan penyaringan (filtering) dengan memberikan kondisi logika di dalam tanda kurung siku `[]`.

Misalnya, kita ingin mencari semua kendaraan listrik yang diproduksi oleh **TESLA** dan berada di kota **Seattle**. Kita juga bisa menggabungkan beberapa kondisi menggunakan operator logika seperti `&` (dan) atau `|` (atau).

```python
df_tesla_only = df_ev[df_ev['Make'] == 'TESLA']
print(f"Jumlah mobil Tesla: {len(df_tesla_only)}")

df_filtered = df_ev[(df_ev['Make'] == 'TESLA') &
                    (df_ev['City'] == 'Seattle') &
                    (df_ev['Model Year'] >= 2022)]

print("\nHasil Filter (Tesla, Seattle, >= 2022):")
display(df_filtered.head())
```

## Renaming Columns

Agar kode kita lebih bersih dan mudah dibaca, kita sering kali perlu mengganti nama kolom yang ada di DataFrame. Misalnya, kita bisa menyingkat nama kolom yang terlalu panjang atau menyeragamkan format penulisannya. Di Pandas, kita menggunakan metode `rename` dengan bantuan *dictionary* untuk memetakan nama kolom lama ke nama kolom yang baru.

```python
df_renamed = df_ev.rename(columns={'Model Year': 'Year',
                                   'Electric Vehicle Type': 'EV_Type'})

print("Data dengan nama kolom yang sudah kita ganti:")
display(df_renamed[['City', 'Make', 'Year', 'EV_Type']].head())
```

## Finding Unique Values and Their Counts

Dalam analisis data, kita sering kali ingin tahu berapa banyak kategori unik yang ada dalam suatu kolom dan berapa frekuensi kemunculannya. Hal ini sangat membantu kita untuk memahami distribusi data—misalnya, merek mobil apa yang paling banyak digunakan di dataset ini. Kita bisa menggunakan fungsi `value_counts()` untuk mendapatkan ringkasan ini secara instan.

```python
print("Top 10 Merek Mobil Listrik Terbanyak:")
print(df_ev['Make'].value_counts().head(10))

print("\nTop 5 Kota dengan Mobil Listrik Terbanyak:")
print(df_ev['City'].value_counts().head(5))
```

## Handling Missing Values

Data yang hilang adalah masalah umum dalam analisis data. Kita punya dua strategi utama untuk mengatasinya:
1. **Menghapus (Deleting):** Jika jumlah data yang hilang sedikit, kita bisa menghapus baris tersebut.
2. **Mengisi (Imputing):** Kita bisa mengisi nilai yang kosong dengan nilai lain, misalnya rata-rata (mean), median, atau nilai yang paling sering muncul (mode).

Di Pandas, kita menggunakan `isnull()` untuk mendeteksi data kosong dan `dropna()` atau `fillna()` untuk menanganinya.

```python
print("Jumlah data kosong per kolom:")
print(df_ev.isnull().sum())

df_clean = df_ev.dropna()
df_filled = df_ev.copy()
df_filled['Legislative District'] = df_filled['Legislative District'].fillna(0)

print(f"\nJumlah baris sebelum dibersihkan: {len(df_ev)}")
print(f"Jumlah baris setelah dropna: {len(df_clean)}")
```
